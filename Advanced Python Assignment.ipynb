```python
'''Module 4: Individual Programming Assignment 1

Parsing Data

This assignment covers your ability to manipulate data in Python.
'''

def relationship_status(from_member="@joaquin",to_member="@jobenilagan",social_graph = {
    "@bongolpoc":{"first_name":"Joselito",
                  "last_name":"Olpoc",
                  "following":[
                  ]
    },
    "@joaquin":  {"first_name":"Joaquin",
                  "last_name":"Gonzales",
                  "following":[
                      "@chums","@jobenilagan"
                  ]
    },
    "@chums" : {"first_name":"Matthew",
                "last_name":"Uy",
                "following":[
                    "@bongolpoc","@miketan","@rudyang","@joeilagan"
                ]
    },
    "@jobenilagan":{"first_name":"Joben",
                   "last_name":"Ilagan",
                   "following":[
                    "@eeebeee","@joeilagan","@chums","@joaquin"
                   ]
    },
    "@joeilagan":{"first_name":"Joe",
                  "last_name":"Ilagan",
                  "following":[
                    "@eeebeee","@jobenilagan","@chums"
                  ]
    },
    "@eeebeee":  {"first_name":"Elizabeth",
                  "last_name":"Ilagan",
                  "following":[
                    "@jobenilagan","@joeilagan"
                  ]
    },
}):

    from_status = social_graph.get(from_member, {}).get("following", [])
    to_status = social_graph.get(to_member, {}).get("following", [])

    if to_member in from_status and from_member in to_status:
        return "friends"
    elif to_member in from_status:
        return "follower"
    elif from_member in to_status:
        return "followed by"
    else:
        return "no relationship"
 
def tic_tac_toe(board=
[
['O','X','X'],
['O','','O'],
['X','X',''],
]
):

    for line in board + list(zip(*board)) + [[board[i][i] for i in range(len(board))], [board[i][len(board) - 1 - i] for i in range(len(board))]]:
        if len(set(line)) == 1 and line[0] != ' ':
            return line[0]

    return "NO WINNER"



def eta(first_stop="admu", second_stop="upd", route_map= {
     ("upd","admu"):{
         "travel_time_mins":60
     },
     ("admu","dlsu"):{
         "travel_time_mins":100
     },
     ("dlsu","upd"):{
         "travel_time_mins":45
     }
}):
    
    current_stop = first_stop
    total_time = 0

    while current_stop != second_stop:
        route = (current_stop, second_stop)
        if route in route_map:
            total_time += route_map[route]["travel_time_mins"]
            break
        next_stops = [key[1] for key in route_map if key[0] == current_stop]
        if not next_stops:
            break 
        next_stop = next_stops[0]
        total_time += route_map[(current_stop, next_stop)]["travel_time_mins"]
        current_stop = next_stop
    return total_time


```
